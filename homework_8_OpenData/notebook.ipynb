```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = [12, 6]
sns.set()
```

```python
dataframe = pd.read_csv("./U.S._Chronic_Disease_Indicators__CDI_.csv", sep=",")
dataframe.dropna().head()
```

```python
dataframe.shape
```

```python
petal_width = dataframe["petal_width"]

print("\nPetal Stat:\n\n", petal_width.describe(), "\n", sep="")

ax = sns.histplot(petal_width, bins=len(petal_width.unique()))
_ = ax.set(xlabel="X", ylabel="Y")
```

```python
data = dataframe[["sepal_length", "sepal_width"]]

plt.suptitle("sepal_length or sepal width")
ax = sns.countplot(data=data, y="sepal_length", hue="sepal_width", alpha = 0.8)
```
